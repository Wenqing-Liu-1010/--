```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_classification_datasets(data_paths):
    """
    绘制多个分类数据集的散点图。

    参数：
    data_paths (list): 数据文件路径列表，每个路径对应一个数据集。
    """
    plt.figure(figsize=(12, 6))

    for i, data_path in enumerate(data_paths):
        # 读取数据
        data = pd.read_csv(data_path, delim_whitespace=True, header=None, names=["x1", "x2", "y"])
        
        # 绘制散点图
        plt.subplot(1, len(data_paths), i + 1)
        sns.scatterplot(data=data, x='x1', y='x2', hue='y', style='y', palette='Set1')
        plt.title(f'Data Set {chr(65 + i)}')  # A, B, C
        plt.xlabel('x1')
        plt.ylabel('x2')

    plt.tight_layout()
    plt.show()

# 使用示例
data_files = [
    "Classification Data for Exercise 6-20241015/classificationA.train",
    "Classification Data for Exercise 6-20241015/classificationB.train",
    "Classification Data for Exercise 6-20241015/classificationC.train"
]

plot_classification_datasets(data_files)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import seaborn as sns

def plot_lda_classification_boundary(data_path):
    """
    绘制 LDA 分类边界。

    参数：
    data_path (str): 数据文件的路径，文件应包含 x1, x2, y 三列。
    """
    data = pd.read_csv(data_path, delim_whitespace=True, header=None, names=["x1", "x2", "y"])

    X = data[["x1", "x2"]]
    y = data["y"]

    lda = LinearDiscriminantAnalysis()
    lda.fit(X, y)

    w = lda.coef_[0]  # 权重
    b = lda.intercept_[0]  # 截距

    # 定义边界函数
    def LDA_curve(x):
        return (-w[0] * x + b) / w[1]

    # 绘制数据点和分类边界
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x='x1', y='x2', hue='y', palette='Set1')
    x_values = np.linspace(data.x1.min(), data.x1.max(), 100)
    plt.plot(x_values, LDA_curve(x_values), color='black', label='LDA Boundary')
    plt.title(f'LDA Classification Boundary for {data_path.split("/")[-1]}')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid()
    plt.show()

# 使用示例
data_files = [
    "Classification Data for Exercise 6-20241015/classificationA.train",
    "Classification Data for Exercise 6-20241015/classificationB.train",
    "Classification Data for Exercise 6-20241015/classificationC.train"
]

for file in data_files:
    plot_lda_classification_boundary(file)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import seaborn as sns

def plot_qda_classification_boundary(data_path):
    """
    绘制 QDA 分类边界。

    参数：
    data_path (str): 数据文件的路径，文件应包含 x1, x2, y 三列。
    """
    # 读取数据集
    data = pd.read_csv(data_path, delim_whitespace=True, header=None, names=["x1", "x2", "y"])

    # 分离特征和标签
    X = data[["x1", "x2"]]
    y = data["y"]

    # 应用 QDA
    qda = QuadraticDiscriminantAnalysis()
    qda.fit(X, y)

    # 创建网格以评估 QDA 分类
    x_min, x_max = data.x1.min() - 1, data.x1.max() + 1
    y_min, y_max = data.x2.min() - 1, data.x2.max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    
    # 预测网格点的类标签
    Z = qda.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # 绘制数据点和 QDA 边界
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x='x1', y='x2', hue='y', palette='Set1')
    plt.contour(xx, yy, Z, levels=[0], colors='black', linewidths=2)  # 绘制边界
    plt.title(f'QDA Classification Boundary for {data_path.split("/")[-1]}')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid()
    plt.show()

# 使用示例
data_files = [
    "Classification Data for Exercise 6-20241015/classificationA.train",
    "Classification Data for Exercise 6-20241015/classificationB.train",
    "Classification Data for Exercise 6-20241015/classificationC.train"
]

for file in data_files:
    plot_qda_classification_boundary(file)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import seaborn as sns

def plot_logistic_regression_boundary(data_path):
    """
    绘制逻辑回归分类边界，并打印系数和截距。

    参数：
    data_path (str): 数据文件的路径，文件应包含 x1, x2, y 三列。
    """
    # 读取数据集
    data = pd.read_csv(data_path, delim_whitespace=True, header=None, names=["x1", "x2", "y"])

    # 分离特征和标签
    X = data[["x1", "x2"]]
    y = data["y"]

    # 应用逻辑回归
    log_reg = LogisticRegression()
    log_reg.fit(X, y)

    # 获取逻辑回归系数和截距
    m1 = log_reg.coef_[0][0]  # x1 的系数
    m2 = log_reg.coef_[0][1]  # x2 的系数
    mc = log_reg.intercept_[0]  # 截距

    # 打印系数和截距
    print(f"Data Set: {data_path.split('/')[-1]}")
    print(f"Coefficient of x1 (m1): {m1}")
    print(f"Coefficient of x2 (m2): {m2}")
    print(f"Intercept (mc): {mc}")
    print()  # 空行分隔不同数据集的输出

    # 定义边界函数
    def log_curve(x):
        return -(mc + m1 * x) / m2

    # 绘制数据点和逻辑回归边界
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=data, x='x1', y='x2', hue='y', palette='Set1')
    x_values = np.linspace(data.x1.min(), data.x1.max(), 100)
    plt.plot(x_values, log_curve(x_values), color='black', label='Logistic Regression Boundary')
    plt.title(f'Logistic Regression Boundary for {data_path.split("/")[-1]}')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.grid()
    plt.show()

# 使用示例
data_files = [
    "Classification Data for Exercise 6-20241015/classificationA.train",
    "Classification Data for Exercise 6-20241015/classificationB.train",
    "Classification Data for Exercise 6-20241015/classificationC.train"
]

for file in data_files:
    plot_logistic_regression_boundary(file)
```

```python
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def load_data(train_path, test_path):
    """
    加载训练和测试数据。
    
    参数：
    train_path (str): 训练数据文件路径。
    test_path (str): 测试数据文件路径。
    
    返回：
    X_train, y_train, X_test, y_test
    """
    train_data = pd.read_csv(train_path, delim_whitespace=True, header=None, names=["x1", "x2", "y"])
    test_data = pd.read_csv(test_path, delim_whitespace=True, header=None, names=["x1", "x2", "y"])
    
    X_train = train_data[["x1", "x2"]]
    y_train = train_data["y"]
    X_test = test_data[["x1", "x2"]]
    y_test = test_data["y"]
    
    return X_train, y_train, X_test, y_test

def compute_classification_metrics(X_train, y_train, X_test, y_test):
    """
    计算每种方法的分类指标，包括准确率、精确率、召回率、F1 分数和 AUC-ROC。
    
    参数：
    X_train (DataFrame): 训练特征数据。
    y_train (Series): 训练标签。
    X_test (DataFrame): 测试特征数据。
    y_test (Series): 测试标签。
    
    返回：
    metrics (dict): 每种方法的分类指标。
    """
    metrics = {}
    
    # LDA
    lda = LinearDiscriminantAnalysis()
    lda.fit(X_train, y_train)
    y_pred_lda = lda.predict(X_test)
    metrics['LDA'] = {
        'Accuracy': accuracy_score(y_test, y_pred_lda),
        'Precision': precision_score(y_test, y_pred_lda),
        'Recall': recall_score(y_test, y_pred_lda),
        'F1 Score': f1_score(y_test, y_pred_lda),
        'AUC': roc_auc_score(y_test, lda.predict_proba(X_test)[:, 1])
    }
    
    # QDA
    qda = QuadraticDiscriminantAnalysis()
    qda.fit(X_train, y_train)
    y_pred_qda = qda.predict(X_test)
    metrics['QDA'] = {
        'Accuracy': accuracy_score(y_test, y_pred_qda),
        'Precision': precision_score(y_test, y_pred_qda),
        'Recall': recall_score(y_test, y_pred_qda),
        'F1 Score': f1_score(y_test, y_pred_qda),
        'AUC': roc_auc_score(y_test, qda.predict_proba(X_test)[:, 1])
    }

    # Logistic Regression
    log_reg = LogisticRegression()
    log_reg.fit(X_train, y_train)
    y_pred_log_reg = log_reg.predict(X_test)
    metrics['Logistic Regression'] = {
        'Accuracy': accuracy_score(y_test, y_pred_log_reg),
        'Precision': precision_score(y_test, y_pred_log_reg),
        'Recall': recall_score(y_test, y_pred_log_reg),
        'F1 Score': f1_score(y_test, y_pred_log_reg),
        'AUC': roc_auc_score(y_test, log_reg.predict_proba(X_test)[:, 1])
    }
    
    return metrics

# 数据集文件路径
data_paths = {
    "A": {
        "train": "Classification Data for Exercise 6-20241015/classificationA.train",
        "test": "Classification Data for Exercise 6-20241015/classificationA.test"
    },
    "B": {
        "train": "Classification Data for Exercise 6-20241015/classificationB.train",
        "test": "Classification Data for Exercise 6-20241015/classificationB.test"
    },
    "C": {
        "train": "Classification Data for Exercise 6-20241015/classificationC.train",
        "test": "Classification Data for Exercise 6-20241015/classificationC.test"
    }
}

# 计算每个数据集的分类指标
results = {}

for key, paths in data_paths.items():
    X_train, y_train, X_test, y_test = load_data(paths['train'], paths['test'])
    results[key] = compute_classification_metrics(X_train, y_train, X_test, y_test)

# 输出结果
for dataset, metrics in results.items():
    print(f"Dataset {dataset}:")
    for method, scores in metrics.items():
        print(f"  {method}:")
        for metric, score in scores.items():
            print(f"    {metric}: {score:.4f}")
```
