```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Load the training data with the correct separator
train_data = pd.read_csv('/Users/wenqing_liu/Desktop/第三学期/Statistical ML/作业/作业1/Data for the pratical application exercises-20241022/simreg1_train.csv', sep=';')

# Extract features and target variable
X_train = train_data['X'].values.reshape(-1, 1)  # Feature variable
y_train = train_data['Y'].values

# Step 2: Load the test data
test_data = pd.read_csv('/Users/wenqing_liu/Desktop/第三学期/Statistical ML/作业/作业1/Data for the pratical application exercises-20241022/simreg1_test.csv', sep=';')
X_test = test_data['X'].values.reshape(-1, 1)  # Feature variable for test set
y_test = test_data['Y'].values  # Target variable for test set

# Step 3: Create the design matrix with the new feature x^2
X_train_poly = np.hstack((np.ones((X_train.shape[0], 1)), X_train, X_train**2))  # Adding bias term and x^2
X_test_poly = np.hstack((np.ones((X_test.shape[0], 1)), X_test, X_test**2))

# Step 4: Fit the linear regression model
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Step 5: Predict values for training and testing sets
y_train_pred = model.predict(X_train_poly)
y_test_pred = model.predict(X_test_poly)

# Step 6: Calculate Mean Squared Error (MSE) for both sets
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print(f'Mean Squared Error (MSE) on Training Set: {mse_train}')
print(f'Mean Squared Error (MSE) on Testing Set: {mse_test}')

# Step 7: Plot the results for visualization
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_train, y_train_pred, color='red', label='Predicted (Train)', alpha=0.5)
plt.title('Polynomial Regression Fit with x^2 Feature')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 1: Load the training data with the correct separator
train_data = pd.read_csv('/Users/wenqing_liu/Desktop/第三学期/Statistical ML/作业/作业1/Data for the pratical application exercises-20241022/simreg1_train.csv', sep=';')

# Extract features and target variable
X_train = train_data['X'].values.reshape(-1, 1)  # Feature variable
y_train = train_data['Y'].values

# Step 2: Load the test data
test_data = pd.read_csv('/Users/wenqing_liu/Desktop/第三学期/Statistical ML/作业/作业1/Data for the pratical application exercises-20241022/simreg1_test.csv', sep=';')
X_test = test_data['X'].values.reshape(-1, 1)  # Feature variable for test set
y_test = test_data['Y'].values  # Target variable for test set

# Step 3: Initialize lists to store MSE values for each k
mse_train_list = []
mse_test_list = []

# Step 4: Loop through k = 1 to 10
for k in range(1, 11):
    # Create the design matrix with features up to x^k
    X_train_poly = np.hstack([X_train ** i for i in range(1, k + 1)])  # Create polynomial features for training set
    X_test_poly = np.hstack([X_test ** i for i in range(1, k + 1)])  # Create polynomial features for testing set
    
    # Step 5: Fit the linear regression model
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Step 6: Predict values for training and testing sets
    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)

    # Step 7: Calculate Mean Squared Error (MSE) for both sets
    mse_train = mean_squared_error(y_train, y_train_pred)+1
    mse_test = mean_squared_error(y_test, y_test_pred)+1

    # Store MSE values
    mse_train_list.append(mse_train)
    mse_test_list.append(mse_test)

    # Step 8: Plot the results for testing set
    plt.figure(figsize=(10, 5))
    plt.scatter(X_test, y_test, color='green', label='Testing Data')
    plt.scatter(X_test, y_test_pred, color='orange', label=f'Predicted (k={k})', alpha=0.5)
    plt.title(f'Polynomial Regression Fit with x^1 to x^{k} (Testing Set)')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid()
    plt.show()

# Step 8: Plot the empirical risks against k
plt.plot(range(1, 11), np.log(mse_test_list), marker='o', label='Test Set MSE', color='orange')
plt.plot(range(1, 11), np.log(mse_train_list), marker='o', label='Train Set MSE', color='blue')
plt.title('Empirical Risks vs. k')
plt.xlabel('k (Degree of Polynomial Features)')
plt.ylabel('Mean Squared Error (MSE)')
plt.xticks(range(1, 11))
plt.legend()
plt.grid()
plt.show()

# Print MSE values for inspection
for k in range(1, 11):
    print(f'k = {k}: Training MSE = {mse_train_list[k-1]}, Testing MSE = {mse_test_list[k-1]}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pinball_function(z, tau):
    """
    Calculate the pinball function h_tau(z) for a given z and tau.
    
    Parameters:
    z : float or np.ndarray
        The input value(s) for which to compute h_tau.
    tau : float
        The parameter tau in the range (0, 1).
    
    Returns:
    np.ndarray
        The computed values of h_tau(z).
    """
    if not (0 < tau < 1):
        raise ValueError("tau must be in the interval (0, 1)")

    # Calculate h_tau(z)
    h_tau = np.where(z <= 0, -tau * z, (1 - tau) * z)
    return h_tau

# Generate continuous values for z
z_values = np.linspace(-3, 3, 500)  # 500 points between -3 and 3

# Define different tau values to plot
tau_values = [0.1, 0.25, 0.5, 0.75, 0.9]

# Plotting the pinball function for different tau values
plt.figure(figsize=(10, 6))
for tau in tau_values:
    h_tau_values = pinball_function(z_values, tau)
    plt.plot(z_values, h_tau_values, label=f'h_tau(z) for tau={tau}')

plt.title('Pinball Function h_tau(z) for Different Tau Values')
plt.xlabel('z')
plt.ylabel('h_tau(z)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 生成随机数据
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 生成100个随机点
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + 噪声

# 中心化数据
X_centered = X - np.mean(X)
y_centered = y - np.mean(y)

# 拟合线性回归模型
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# 拟合仿射回归模型
model_centered = LinearRegression()
model_centered.fit(X_centered, y_centered)
y_centered_pred = model_centered.predict(X_centered)

# 可视化结果
plt.figure(figsize=(12, 6))

# 原始数据与线性回归
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Linear regression')
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()

# 中心化数据与仿射回归
plt.subplot(1, 2, 2)
plt.scatter(X_centered, y_centered, color='green', label='Centered data points')
plt.plot(X_centered, y_centered_pred, color='orange', label='Affine regression')
plt.title('Affine Regression (Centered Data)')
plt.xlabel('Centered X')
plt.ylabel('Centered y')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

# 输出偏置项
print(f"Linear regression intercept (b): {model.intercept_[0]}")
print(f"Affine regression intercept (b_hat): {model_centered.intercept_[0]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

# 生成数据
np.random.seed(42)
X = np.linspace(0, 1, 10).reshape(-1, 1)  # 10个样本
y = np.sin(2 * np.pi * X) + np.random.normal(0, 0.2, X.shape)  # 添加噪声

# 绘制没有正则化的多项式回归
plt.figure(figsize=(12, 6))

# 不带正则化的多项式回归
plt.subplot(1, 2, 1)
model_poly = make_pipeline(PolynomialFeatures(degree=9), LinearRegression())
model_poly.fit(X, y)
X_test = np.linspace(0, 1, 100).reshape(-1, 1)
y_pred_poly = model_poly.predict(X_test)

plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_test, y_pred_poly, color='red', label='Polynomial Regression (No Regularization)')
plt.title('Polynomial Regression without Regularization')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid()

# 带正则化的岭回归
plt.subplot(1, 2, 2)

# 使用网格搜索选择最佳的正则化参数lambda
alpha_range = np.logspace(-10, -1, 10000)  # 生成 10 个 lambda 值
ridge_model = Ridge()
param_grid = {'alpha': alpha_range}
grid_search = GridSearchCV(ridge_model, param_grid, cv=5)  # 5折交叉验证
model_ridge = make_pipeline(PolynomialFeatures(degree=9), grid_search)
model_ridge.fit(X, y)

# 获取最佳的正则化参数
best_alpha = grid_search.best_params_['alpha']
print(f'最佳正则化参数 (lambda): {best_alpha}')

# 使用最佳参数进行预测
y_pred_ridge = model_ridge.predict(X_test)

plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_test, y_pred_ridge, color='green', label='Ridge Regression')
plt.title('Polynomial Regression with Ridge Regularization')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge, OrthogonalMatchingPursuit
from sklearn.metrics import mean_squared_error

# 设置参数
n = 64  # 样本数量
p_range = [4, 256]  # 特征数量范围 (确保p至少为4以满足非零系数需求)
sigma_squared = 1  # 噪声方差
non_zero_coeffs = 4  # 非零系数数量

# 用于存储结果
dimensions = range(p_range[0], p_range[1] + 1)
mse_lasso = []
mse_ridge = []
mse_omp = []

# 迭代不同的维度
for p in dimensions:
    # 生成高斯设计矩阵 X
    X = np.random.randn(n, p)
    
    # 确保特征数量至少为非零系数数量
    if p < non_zero_coeffs:
        continue  # 跳过该迭代

    # 随机选择 4 个非零系数的索引
    true_coeffs = np.zeros(p)
    non_zero_indices = np.random.choice(p, non_zero_coeffs, replace=False)
    true_coeffs[non_zero_indices] = np.random.choice([-1, 1], size=non_zero_coeffs) * 4  # 非零系数取值为 -4 或 4
    
    # 生成目标变量 y
    y = X @ true_coeffs + np.random.normal(0, np.sqrt(sigma_squared), n)  # 添加噪声
    
    # Lasso回归
    lasso = Lasso(alpha=0.1)
    lasso.fit(X, y)
    y_pred_lasso = lasso.predict(X)
    mse_lasso.append(mean_squared_error(y, y_pred_lasso))
    
    # Ridge回归
    ridge = Ridge(alpha=0.1)
    ridge.fit(X, y)
    y_pred_ridge = ridge.predict(X)
    mse_ridge.append(mean_squared_error(y, y_pred_ridge))
    
    # OMP回归
    omp = OrthogonalMatchingPursuit(n_nonzero_coefs=non_zero_coeffs)
    omp.fit(X, y)
    y_pred_omp = omp.predict(X)
    mse_omp.append(mean_squared_error(y, y_pred_omp))

# 绘制结果
plt.figure(figsize=(10, 6))
plt.plot(dimensions, mse_lasso, marker='o', label='Lasso')
plt.plot(dimensions, mse_ridge, marker='o', label='Ridge')
plt.plot(dimensions, mse_omp, marker='o', label='OMP')
plt.xscale('linear')
plt.xlabel('Number of Features (p)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('MSE Comparison of Lasso, Ridge, and OMP')
plt.legend()
plt.grid()
plt.show()
```
